### 📌4. CHURN ANALYSIS

```python
churned_users = df[df["is_churned"] == 1]
churned_users
```

```python
df["is_churned"].value_counts()
```

- Churn means when a user stops using a service or cancels their subscription.In Spotify’s case, churned users are those who either Cancel their premium subscription, or Stop being active on the platform for a certain period. It’s basically the opposite of retention.
Churn analysis is the process of studying why users leave and identifying patterns that predict churn. It helps companies reduce user loss and improve retention.
- User churn status (1 = Churned, 0 = Active)

#### SUBSCRIPTION TYPE VS CHURN :

```python
plt.figure(figsize = (10,5))
sns.countplot(x = "is_churned", hue = "subscription_type", data = df, palette = "Set1")
plt.title("Churn by Subscription Type", fontsize = 14)
plt.xlabel("Subscription Type", labelpad = 15)
plt.ylabel("User Count", labelpad = 15)
plt.show()
```

📊KEY INSIGHTS :
- Retention is Dominated by "Free" Users : The Free plan has the highest count among users who have not churned. This suggests a large base of non-paying users who are relatively stable, likely because there is no financial cost to maintaining the account.
- High Churn Risk in "Premium" and "Family" : When looking at the churned group, the Premium and Family segments show significant representation. While the total volume of retained users is high, the proportion of paid subscribers leaving is a critical concern for revenue stability.
- "Student" Plan Stability : The Student tier appears consistently across both categories but shows a very high retention-to-churn ratio. This indicates that the student discount is likely an effective retention tool, keeping younger users in the ecosystem with fewer cancellations.


#### LISTENING TIME VS CHURN :

```python
fig = px.box(df, x = "is_churned", y = "listening_time", color = "subscription_type",
             title = "Listening Time Distribution by Churn Status",
        labels = {"is_churned":"Churned vs Active","listening_time":"Listening Time (minutes/day)"})
fig.show()
```

📊KEY INSIGHTS :
- Active Users (0) : The median listening time for active users (indicated by the horizontal line within the boxes on the left) is generally higher across all subscription types compared to churned users. This suggests that sustained higher daily listening time is a strong indicator of user retention.
- Churned Users (1) : Churned users exhibit a lower median and a more compressed interquartile range (the box itself), which indicates lower overall engagement before they discontinued the service.
- Premium & Student Tiers : These groups typically show higher total listening minutes. However, research indicates that the Student segment can be particularly volatile; they may also have higher churn risks due to budget sensitivity or lifestyle changes.
- Free Tier : This segment often has the widest spread of listening times but is highly susceptible to churn if ad exposure becomes excessive, leading to lower engagement patterns.

#### CHURN STATUS BY SKIP RATE :

```python
plt.figure(figsize=(8,5))
sns.kdeplot(x = "is_churned", y = "skip_rate", data = df)
plt.title("Churn Status by Skip Rate", fontsize = 14)
plt.xlabel("Churned (1) VS Active (0)", labelpad = 15)
plt.ylabel("Skip Rate")
plt.show()
```

📊KEY INSIGHTS :
- Strong Correlation with Churn : There is a clear distinction between the skip rates of active users versus churned users. Higher skip rates are a strong indicator of potential churn.

- Churned Users (1) : The density for churned users is heavily concentrated at the top of the scale, peaking between 0.8 and 1.0. This indicates that users who eventually leave the service have very high skip behaviors.
- Predictive Value : Because the two clusters are so visually distinct with minimal overlap, "Skip Rate" is likely one of the most important features for building a predictive churn model.

### 📌5. CORRELATION ANALYSIS

```python
corr = df.select_dtypes(include = "number").corr()
corr
```

- Correlation analysis is a statistical method used to measure the strength and direction of the relationship between two variables.
- Positive correlation → both variables increase together (e.g., listening time ↑, retention ↑).
- Negative correlation → one increases while the other decreases (e.g., ads ↑, retention ↓).
- No correlation → variables don’t show a consistent relationship.
- The most common measure is the Pearson correlation coefficient (r), which ranges from -1 to +1:
    - +1 → perfect positive relationship
    - -1 → perfect negative relationship
    -  0 → no relationship

#### CORRELATION MATRIX - KEY VARIABLES :

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot = True, cmap = "coolwarm")
plt.title("Correlation Matrix", fontsize = 14)
plt.xticks(rotation = 75)
plt.show()
```

👉🏻VISUAL INSIGHTS :
- A correlation matrix is a table showing correlation coefficients between multiple variables at once.
- Rows and columns represent variables.
- Each cell shows the correlation value between two variables.
- Diagonal values are always 1 (a variable perfectly correlates with itself).

📊KEY INSIGHTS :
- Strong Negative Correlation(-0.88) : There is a significant inverse relationship between offline_listening and ads_listened_per_week. This suggests that users who listen to more music offline encounter far fewer advertisements, likely indicating they are on a premium or paid tier.
- Negligible Correlation with Churn : The target variable, is_churned, shows very weak correlations with all other features (ranging from -0.016 to 0.016). This indicates that no single metric in this list is a 'strong linear predictor' of whether a user will stop using the service.
- Feature Independence : Most features (like age, listening_time, and skip_rate) show correlations near zero with one another. This suggests these variables provide unique, non-redundant information for any further data modeling.
